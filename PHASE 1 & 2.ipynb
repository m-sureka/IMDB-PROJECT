```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r'/Users/muskansureka/Desktop/chromedriver.exe')
```

```python
driver.get('https://www.imdb.com/search/title/?genres=action&sort=user_rating,desc&title_type=feature&num_votes=25000,&pf_rd_m=A2FGELUUNOQJNL&pf_rd_p=f11158cc-b50b-4c4d-b0a2-40b32863395b&pf_rd_r=XZ8X52H1R40B7KG5SNZ9&pf_rd_s=right-6&pf_rd_t=15506&pf_rd_i=top&ref_=chttp_gnr_1')
```

```python
Sno=[]
start=0
end=2
for sno in range(start,end):
    snos=driver.find_elements(By.XPATH,'//span[@class="lister-item-index unbold text-primary"]')
    for i in snos:
        Sno.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Year=[]
start=0
end=2
for year in range(start,end):
    years=driver.find_elements(By.XPATH,'//span[@class="lister-item-year text-muted unbold"]')
    for i in years:
        Year.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Duration=[]
start=0
end=2
for duration in range(start,end):
    durations=driver.find_elements(By.XPATH,'//span[@class="runtime"]')
    for i in durations:
        Duration.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Rating=[]
start=0
end=2
for rating in range(start,end):
    ratings=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-imdb-rating"]')
    for i in ratings:
        Rating.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
Metascore=[]
start=0
end=2
for metascore in range(start,end):
    try:
        metascores=driver.find_elements(By.XPATH,'//div[@class="inline-block ratings-metascore"]//span')
        for i in metascores:
            Metascore.append(i.text)
    except NoSuchElementException:
        Metascore.append('-')
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Votes=[]
start=0
end=2
for vote in range(start,end):
    votes=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in votes:
        Votes.append(i.text[7:16])
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Gross=[]
start=0
end=2
for gross in range(start,end):
    grosss=driver.find_elements(By.XPATH,'//p[@class="sort-num_votes-visible"]')
    for i in grosss:
        Gross.append(i.text[26:34])
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Genre=[]
start=0
end=2
for genre in range(start,end):
    genres=driver.find_elements(By.XPATH,'//span[@class="genre"]')
    for i in genres:
        Genre.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Certification=[]
start=0
end=2
for certificate in range(start,end):
    certificates=driver.find_elements(By.XPATH,'//span[@class="certificate"]')
    for i in certificates:
        Certification.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Movie=[]
start=0
end=2
for movie in range(start,end):
    movies=driver.find_elements(By.XPATH,'//h3[@class="lister-item-header"]//a')
    for i in movies:
        Movie.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```

```python
Director=[]
start=0
end=2
for director in range(start,end):
    directors=driver.find_elements(By.XPATH,'//p[3][@class]/a[1]')
    for i in directors:
        Director.append(i.text)
    next_button=driver.find_element(By.XPATH,'//*[@id="main"]/div/div[1]/div[2]/a')
    next_button.click()
    time.sleep(2)
```
